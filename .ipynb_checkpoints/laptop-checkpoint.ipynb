# <font color=2E8B57> *Laptop Price Prediction*

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('../input/laptop-price-predictor/laptop_pricing.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df2=pd.get_dummies(df.Manufacturer,prefix='Manufacturer')
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df=pd.concat([df,df2],axis='columns')
```

```python
df.head()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(numeric_only=True),annot=True);
```

```python
df.drop('Manufacturer',inplace=True,axis='columns')
```

```python
x=df.drop('Price',axis='columns')
y=df['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
r2_score(y_test,pred)
```

```python
rf=RandomForestRegressor(random_state=0)
```

```python
rf.fit(x, y)
```

```python
pred2=rf.predict(x_test)
```

```python
r2_score(y_test,pred2)
```
